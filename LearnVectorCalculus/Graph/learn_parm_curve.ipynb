```python
import numpy as np 
import matplotlib.pyplot as plt 


t = np.arange(0, 2*np.pi, 0.1)

x = np.sin(t)
y = 4*np.cos(t)

plt.plot(x, y)
plt.show()
```

```python
fig = plt.figure()

ax = fig.gca(projection='3d')

t = np.arange(-10, 10, 0.1)

x = 2*t - 1
y = t + 2
z = t 

ax.plot(x, y, z)

plt.show()
```
